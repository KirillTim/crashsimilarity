```python
vocabs = dict([(i, pickle.load(open('data/model-size/vocab_top_{}.pickle'.format(i), 'rb'))) for i in [10,20,30]])
```

```python
corpus1 = pickle.load(open('data/model-size/multiple_sig_last_year.pickle', 'rb'))
corpus1 = [i['prepared'] for i in corpus1]
words1 = set()
for group in corpus1:
    for g in group:
        for w in g:
            words1.add(w)
len(words1)
```

```python
corpus2 = pickle.load(open('data/model-size/multiple_sig_year_before.pickle', 'rb'))
corpus2 = [i['prepared'] for i in corpus2]
words2 = set()
for group in corpus2:
    for g in group:
        for w in g:
            words2.add(w)
len(words2)
```

```python
cnt2 = dict([(i, 0) for i in vocabs.keys()])
for w in words2:
    for i in cnt2.keys():
        if w in vocabs[i].keys():
            cnt2[i] +=1 
for (k,v) in cnt2.items():
    cnt2[k] = v / float(len(words2))
cnt2
```
